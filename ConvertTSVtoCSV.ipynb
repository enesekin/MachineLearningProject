```python
import re

# reading given tsv file
with open("/content/gdrive/MyDrive/MachineLearningProject/amazon_reviews_us_Musical_Instruments_v1_00.tsv", 'r') as myfile:  
  with open("/content/gdrive/MyDrive/MachineLearningProject/amazon_reviews_us.csv", 'w') as csv_file:
    for line in myfile:
      # Replace every tab with comma
      fileContent = re.sub("\t", ",", line)
        
      # Writing into csv file
      csv_file.write(fileContent)
  
# output
print("Successfully made csv file")
```
